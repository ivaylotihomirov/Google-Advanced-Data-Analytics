Next, convert the categorical features to indicator (one-hot encoded) features. 

**Note:** The `drop_first` argument can be kept as default (`False`) during one-hot encoding for random forest models, so it does not need to be specified. Also, the target variable, `satisfaction`, does not need to be encoded and will be extracted in a later step.

```python
# Convert categorical features to one-hot encoded features.

### YOUR CODE HERE ###

cat_cols = air_data_subset.select_dtypes(include=['object', 'category']).columns.tolist()

# 2. Remove the specific column you want to exclude
exclude_col = 'satisfaction'
if exclude_col in cat_cols:
    cat_cols.remove(exclude_col)

# 3. Convert the remaining columns
air_data_subset_dummies = pd.get_dummies(air_data_subset, columns=cat_cols, drop_first=True)
```

The first step to building your model is separating the labels (y) from the features (X).

### Tune the model

Now, fit and tune a random forest model with separate validation set. Begin by determining a set of hyperparameters for tuning the model using GridSearchCV.


```python
# Determine set of hyperparameters.

### YOUR CODE HERE ###

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

Next, create a list of split indices.

```python
# Create list of split indices.

### YOUR CODE HERE ###

X_tr['is_val'] = -1
X_val['is_val'] = 0

# 2. Combine them into one unified DataFrame
X_combined = pd.concat([X_tr, X_val], axis=0).reset_index(drop=True)
y_combined = pd.concat([y_tr, y_val], axis=0).reset_index(drop=True)

# 3. Extract the split array and drop it from the features
# (We don't want the model to accidentally train on our split indicator!)
split_indices = X_combined['is_val'].values
X_combined = X_combined.drop(columns=['is_val'])

pds = PredefinedSplit(test_fold=split_indices)
```

```python
# Instantiate model.

### YOUR CODE HERE ### 

rf = RandomForestClassifier(random_state=42)
```
